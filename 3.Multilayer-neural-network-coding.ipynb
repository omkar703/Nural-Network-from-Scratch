#### **Multi-layer neural network coding**

```python
# 2 layers 

inputs = [[1,2,3,2.5],
          [1,2,4,2.5],
          [1,2,3,2]]

weights1 = [[0.2,0.8,-0.5,1],
            [0.5,-0.91,0.26,-0.5],
            [-0.26,-0.27,0.17,1]]

biases1 = [2,3,0.5]

weights2 = [[0.2,0.8,-0.5],
            [0.5,-0.91,0.26],
            [-0.26,-0.27,0.17]]

biases2 = [2,3,0.5]

import numpy as np

outputs1 = np.dot(inputs,np.transpose(weights1)) + biases1
outputs2 = np.dot(outputs1,np.transpose(weights2)) + biases2

print(outputs2)
```
