#### **Neuron coding using numpy**

```python
inputs = [1,2,3,2.5]
weights = [0.2,0.8,-0.5,1]
biases = 2

import numpy as np

outputs = np.dot(inputs,weights)+biases
print(outputs)

```

#### used on layer

```python
inputs = [1.0,2.0,3.0,2.5]
weights = [[0.2,0.8,-0.5,1],
           [0.5,-0.91,0.26,-0.5],
           [-0.26,-0.27,0.17,1]]

# ( 3 * 4 ) ( 4 * 1) = ( 3 * 1) -> ( 3 * 1 ) + ( 1 * 3 ) = ( 3 * 1 )
biases = [2,3.5,1]


layer_outputs = np.dot(weights,inputs) + biases
print(layer_outputs)

```

```python
# batch of inputs 
inputs  = [[1,2,3,0.5],
           [1,2,4,2.5],
           [1,2,3,2]]

weights = [[0.2,0.8,-0.5,1],
           [0.5,-0.91,0.26,-0.5],
           [-0.26,-0.27,0.17,1]]

biases = [2,3,0.5]

import numpy as np



outputs = np.dot(inputs,np.transpose(weights)) + biases
print(outputs)

```
